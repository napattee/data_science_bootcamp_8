```python
!pip install gazpacho
import requests
from gazpacho import Soup
import pandas as pd

url_boxoffice = "https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW"

html = requests.get(url_boxoffice)

boxoffice = Soup(html.text)
titles = boxoffice.find("td", {"class": "a-text-left mojo-field-type-title"})
grosses = boxoffice.find("td", {"class": "a-text-right mojo-field-type-money"})
ranking = boxoffice.find("td", {"class": "a-text-right mojo-header-column mojo-truncate mojo-field-type-rank"})
years = boxoffice.find("td", {"class": "a-text-left mojo-field-type-year"})

clean_titles = [title.strip() for title in titles]
clean_grosses = [gross.strip() for gross in grosses]
clean_ranking = [int(rank.strip()) for rank in ranking]
clean_years = [int(year.strip()) for year in years]

clean_years
len(clean_years)

new_clean_gross = [int(newgross.replace("$", "").replace(",", "")) for newgross in clean_grosses]

boxoffice_database = pd.DataFrame(data = {
    "rank" : clean_ranking,
    "title" : clean_titles,
    "year" : clean_years,
    "gross" : new_clean_gross
})
```

```python
boxoffice_database.groupby("year")["gross"].agg(["max", "sum", "mean"]).reset_index()
```
